### Creating a dataframe that contains Nan to deal with it.

```python
data = {
    'Name': ['Alice', None, 'Charlie', 'David', 'Eve', 'Eve'],
    'Age': [24, 27, None, 32, 29, 29],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix','Phoenix']
}
df = pd.DataFrame(data)
df
```

### Data Cleaning

### filling the Nan in age column with the mean value.
### inplace = True means that changes that directly in the original dataframe instead of returning another one.

```python
# Handling missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

### filling the Nan in the name column with the mode "only the first mode" incase there is other modes.

```python
df['Name'].fillna(df['Name'].mode()[0], inplace=True)
df
```

```python
df.dropna(subset=['Name'], inplace=True)
df
```

### checking for duplicates in the data

```python
print(df.duplicated())

df.drop_duplicates(inplace=True)
df
```

### Renaming columns using a dictionary where the keys are the exisiting column name and the value is the new column name.

```python
df.rename(columns={'City': 'Location' }, inplace=True)
df
```

### Data Transformation

### updating column values by 1 using lambda

```python
df['Age'] = df['Age'].apply(lambda x: x + 1)
```

### Binning data
### cut function is used to convert continous values into categorical values by sorting them
### bins list is the edge of each bin
### first bin : 20 to 25
### second bin : greater than 25 to 30 and so on.

```python
df['Age Group'] = pd.cut(df['Age'], bins=[20,25, 30, 40], labels=['20-25','25-30' ,'30-40'])
```

```python
df
```

### Encoding categorical variables
### get_dummies function is used to encode the categorical variables

```python
df = pd.get_dummies(df, columns=['Location'])
df
```
